```python
import pandas as pd

data_cleaned  = pd.read_csv( 'Air Quality Prediction Cleaned.csv')
data_cleaned 

df = pd.DataFrame(data_cleaned)
df
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)

df['Date'].head()
```

```python
df.isnull().sum()
```

```python
df['Date'] = df['Date'].ffill()  # Forward fill missing dates
```

```python
df.isnull().sum()
```

```python
df['AQI'] = df['AQI'].fillna(df['AQI'].mean())
```

```python
df['AQI_LVL'] = df['AQI_LVL'].fillna(df['AQI_LVL'].mode()[0])
```

```python
df.isnull().sum()
```

```python
def fill_aqi_level(row):
    if pd.isnull(row['AQI_LVL']): 
        if row['AQI'] <= 50:
            return "Good"
        elif row['AQI'] <= 100:
            return "Satisfactory"
        elif row['AQI'] <= 150:
            return "Moderate"
        elif row['AQI'] <= 200:
            return "Unhealthy"
        elif row['AQI'] <= 300:
            return "Very Unhealthy"

    else:
        return row['AQI_LVL'] 

df['AQI_LVL'] = df.apply(fill_aqi_level, axis=1)
```

```python
df.isnull().sum()
```
